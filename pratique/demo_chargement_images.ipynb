```python
#Importations des bibliothèques 
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread
from scipy import ndimage
```

```python
#Chargement d'une image
image_url = 'https://www.parolesvivantes.com/nature/eau/eau3.jpg'

img = imread(image_url)
print("Image bien chargée")
print("Dimensions de l'image :", img.shape)
```

```python
#Affichage de l'image en couleur

plt.imshow(img)
plt.title("Image originale")
plt.axis("off") #Permet de retirer l'affichage des axes
plt.show()
```

```python
#Fontion personnalisée pour convertir l'image en niveaux gris
#On aura à utiliser une pondération basée sur la sensibilité de l'oeil humain

def rgb2gray(img) : 
    return 0.299 * img[:,:,0] + 0.587 * img[:,:,1] + 0.114 * img[:,:,2]

gray = rgb2gray(img)

print("Conversion réussie")
print("Dimensions :", gray.shape)
```

```python
#Affichage de l'image en gris
plt.imshow(gray, cmap='gray')
plt.title("Image en gris")
plt.axis("off")
plt.show()
```
